```python
import osmnx as ox
import matplotlib.pyplot as plt
from shapely_polyskel import StraightSkeleton
import shapely.affinity

ox.settings.use_cache = False
ox.settings.log_console = False
```

References:

- [Retrieving OpenStreetMap data](https://pythongis.org/part2/chapter-09/nb/00-retrieving-osm-data.html)

```python
# Get area boundary (geopandas.geodataframe.GeoDataFrame)
place_name = "Pierzchnica, Poland"
area = ox.geocode_to_gdf(place_name).to_crs(3857)

# Get area polygon (shapely.geometry.polygon.Polygon)
geometry = area.get_geometry(0)
polygon = geometry.values[0]

# Changing the location point of the polygon
min_x, min_y, _, _ = polygon.bounds
polygon = shapely.affinity.translate(polygon, -min_x, -min_y)

StraightSkeleton(polygon)
```
